```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):          # Getter
        return self._radius

    @radius.setter
    def radius(self, value):   # Setter
        if value < 0:
            raise ValueError("Radius cannot be negative")
        self._radius = value

c = Circle(10)
print(c.radius)   # ‚úÖ calls getter
c.radius = 20     # ‚úÖ calls setter
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):  # Inherit from ABC
    @abstractmethod
    def area(self):   # Abstract method ‚Äî no implementation
        pass
    #@abstractmethod
    #def circumference(self): #Can't instantiate abstract class Circle without an implementation for abstract method 'circumference'
     #   pass
    def m1(self):
        print("hii")

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):   # Must override abstract method
        return 3.14 * self.r * self.r
    

c = Circle(5)
print(c.area())  # ‚úÖ Works fine
c.m1()
```

```python
from abc import ABC, abstractmethod

class Demo(ABC):
    @property
    @abstractmethod
    def name(self):
        pass

    @classmethod
    @abstractmethod
    def hello(cls):
        pass

    @staticmethod
    @abstractmethod
    def greet():
        pass


class Child(Demo):
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @classmethod
    def hello(cls):
        print(f"Hello from class: {cls.__name__}")

    @staticmethod
    def greet():
        print("Static greeting!")


c = Child("Ankita")
print(c.name)   # property
Child.hello()   # classmethod
c.greet()       # staticmethod
```

Let‚Äôs start simple:

That means:

A is an object created by the metaclass type.

Just like a = A() creates an instance of A,
‚Üí A = type('A', (), {}) creates a class object of type.

```python
class A:
    pass

a = A()
print(isinstance(a, A))      # True
print(isinstance(A, type))   # True
```

```python
class Meta(type):
    def __new__(cls, name, bases, attrs):
        print("Meta __new__ called")
        return super().__new__(cls, name, bases, attrs)

    def __init__(cls, name, bases, attrs):
        print("Meta __init__ called")

class MyClass(metaclass=Meta):
    pass
```

```python
class MyMeta(type):       # must inherit from 'type'
    def __new__(cls, name, bases, attrs):
        print(f"Creating class {name} using {cls}")
        attrs['class_id'] = 1234  # automatically add new class variable
        return super().__new__(cls, name, bases, attrs)

class MyClass(metaclass=MyMeta):
    pass

print(MyClass.class_id)
```

nested class
A class defined inside another class.

It‚Äôs used when one class logically belongs to another and isn‚Äôt needed elsewhere.

üëâ It helps in encapsulation, namespacing, and keeping related logic together.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.address = self.Address("Pune", "Maharashtra")

    def show(self):
        print(self.name, self.age)
        self.address.display()

    class Address:     # Inner Class
        def __init__(self, city, state):
            self.city = city
            self.state = state

        def display(self):
            print(self.city, self.state)

p1 = Person("Ankita", 22)
p1.show()
#The inner class Address is created only within Person.

#You can access it via the outer class:
addr = Person.Address("Mumbai", "MH")
addr.display()
```

Proxy Pattern

```python
class RealServer:
    def access(self):
        print("Accessing real server...")

class ProxyServer:
    def __init__(self):
        self.real_server = RealServer()

    def access(self, user):
        if user == "admin":
            self.real_server.access()
        else:
            print("Access Denied!")

proxy = ProxyServer()
proxy.access("guest")
proxy.access("admin")
```

adaptor pattern

```python
class JSONData:
    def get_data(self):
        return {"name": "Ankita"}

class Adapter:
    def __init__(self, obj):
        self.obj = obj

    def get_data_xml(self):
        data = self.obj.get_data()
        return f"<name>{data['name']}</name>"

json_obj = JSONData()
adapter = Adapter(json_obj)
print(adapter.get_data_xml())  # <name>Ankita</name>
```

Goal: Add new functionality without changing the existing class.

```python
def log_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@log_decorator
def greet(name):
    print("Hello,", name)

greet("Ankita")
```

Observer Pattern

Goal: When one object changes, others should be automatically notified.
Real-world:

YouTube Subscriber system

GUI event listeners

```python
class Subject:
    def __init__(self):
        self._observers = []

    def attach(self, obs): self._observers.append(obs)
    def notify(self, msg):
        for obs in self._observers:
            obs.update(msg)

class Subscriber:
    def update(self, msg):
        print("Got update:", msg)

channel = Subject()
s1 = Subscriber()
s2 = Subscriber()

channel.attach(s1)
channel.attach(s2)
channel.notify("New video uploaded!")
```

Factory Method

Goal: Create objects without exposing creation logic to the client.

üîπ Real-world example:

ShapeFactory returning Circle or Rectangle depending on input.

DatabaseFactory returning MySQL or SQLite connector.

```python
class Shape:
    def draw(self): pass

class Circle(Shape):
    def draw(self): print("Drawing Circle")

class Square(Shape):
    def draw(self): print("Drawing Square")

class ShapeFactory:
    def get_shape(self, shape_type):
        if shape_type == "circle": return Circle()
        elif shape_type == "square": return Square()
        else: return None

factory = ShapeFactory()
shape = factory.get_shape("circle")
shape.draw()
```

Singleton Pattern

Goal: Ensure only one instance of a class exists.

üîπ Real-world example:

Database connection object

Logging system

Configuration manager

```python
class Singleton:
    _instance = None  # shared variable

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance

s1 = Singleton()
s2 = Singleton()
print(s1 is s2)  # ‚úÖ True ‚Äî same object
```
